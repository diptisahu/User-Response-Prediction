```python
import pandas as pd
```

```python
data = pd.read_csv('../data/fm_training.csv')
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
cols = data.columns.tolist()
print(cols)
```

```python
cols = cols[-1:] + cols[:-1]
cols
```

```python
data = data[cols]
```

```python
data.head()
```

```python
# data.to_csv('data/index_first_testing.csv', header = False, index=False)
```

```python
train_x = data.drop(columns=['click'])
train_x.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
principalComponents = pca.fit_transform(train_x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 
                         'principal component 3', 'principal component 4', 
                          'principal component 5', 'principal component 6', 
                          'principal component 7', 'principal component 8', 
                         ])
```

```python
principalDf.head()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
X_train_lda = lda.fit_transform(train_x, data['click'])
# X_test_lda = lda.transform(X_test_std)
X_train_lda.shape
```
